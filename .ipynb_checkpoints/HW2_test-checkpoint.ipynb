```python
# Traversal
# Given only a start node
start = "Luke Gilbert"
        
# Given a start node
queue = [start] # initialize a queue, with the start node
visited = [] # initialize a visited list with the start ndoe

# While the queue is not empty
counter = 0
while queue:
    counter += 1 # just a current counter
    #print("Iteration " + str(counter))
    #print("Current queue list:")
    #print(queue)
    current_node = queue.pop(0) # Get the current node from first in the queue list
    #print("Current node: " + str(current_node))
    visited.append(current_node) # Add the current node to the list of visited
    frontier = [n for n in g.graph.neighbors(current_node)]
    #print("Current frontier list")
    #print(frontier)
    for frontier_node in frontier:
        if frontier_node not in visited and frontier_node not in queue:
            # Add frontier node if it has not yet been visited AND
            # If the frontier node has not yet been added to the queue
            #print("Adding frontier node " + str(frontier_node) + " to queue")
            queue.append(frontier_node)
    #print("Visited list")
    #print(visited)
    #print("")
```

```python
# Given both a start and end node
# Given only a start node
start = "Luke Gilbert"
end = "Martin Kampmann"
        
# Given a start node
queue = [start] # initialize a queue, with the start node
visited = [] # initialize a visited list with the start ndoe

## Add a break so you don't run further than necessary 
break_flag = False
# Keep track of the parent node
parent_node = {start:None}
# While the queue is not empty
while queue:
    current_node = queue.pop(0) # Get the current node from first in the queue list
    visited.append(current_node) # Add the current node to the list of visited
    frontier = [n for n in g.graph.neighbors(current_node)]
    for frontier_node in frontier:
        if frontier_node not in visited and frontier_node not in queue:
            # Add frontier node if it has not yet been visited AND
            # If the frontier node has not yet been added to the queue
            queue.append(frontier_node)
            # Save the current node as the parent node of the frontier node
            parent_node[frontier_node] = current_node
            if frontier_node == end:
                break_flag = True
                break;
        if break_flag:
            break;
```

```python
end = "Martin Kampmann"
#end = "Reza Abbasi-Asl"
# Check to make sure the end node exists in the parent_node
if end not in parent_node:
    print("Does not exist")
else:
    node_lineage = [end] # List to store parent node
    curr_node = end
    while curr_node: # While current node is not equal to none
        # Get parent_node
        nde = parent_node[curr_node]
        # Append the parent node to the node lineage list
        node_lineage.append(nde)
        # Update the current_node
        curr_node = nde

    # Remove the none at the end
    node_lineage.pop()
    node_lineage = node_lineage[::-1] # Reverse order of node lineage
    
```
